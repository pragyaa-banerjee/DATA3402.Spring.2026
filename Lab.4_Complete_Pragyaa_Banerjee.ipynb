1. Write a "counter" class that can be incremented up to a specified maximum value, will print an error if an attempt is made to increment beyond that value, and allows reseting the counter. 

```python
class Counter:
    def __init__(self, maximum):
        self.val = 0
        self.max = maximum

    def increment(self):
        if self.val < self.max:
            self.val += 1
        else:
            print("Error: Counter is already at maximum!")

    def reset(self):
        self.val = 0
```

3. Implement a class to represent a rectangle, holding the length, width, and $x$ and $y$ coordinates of a corner of the object. Implement functions that compute the area and perimeter of the rectangle. Make all data members private and privide accessors to retrieve values of data members. 

```python
class Rectangle:
    def __init__(self, length, width, x, y):
        self.len = length
        self.w = width
        self.x = x
        self.y = y

    def area(self):
        return self.len * self.w

    def perimeter(self):
        return 2 * (self.len + self.w)

    # accessors
    def get_length(self):
        return self.len

    def get_width(self):
        return self.w

    def get_position(self):
        return (self.x, self.y)
```

```python
r = Rectangle(10, 5, 2, 3)

print("Area:", r.area())
print("Perimeter:", r.perimeter())
print("Position:", r.get_position())
```

4. Implement a class to represent a circle, holding the radius and $x$ and $y$ coordinates of center of the object. Implement functions that compute the area and perimeter of the rectangle. Make all data members private and privide accessors to retrieve values of data members. 

```python
import math

class Circle:
    def __init__(self, radius, x, y):
        self.rad = radius
        self.x = x
        self.y = y

    def area(self):
        return math.pi * self.rad ** 2

    def perimeter(self):
        return 2 * math.pi * self.rad

    def get_radius(self):
        return self.rad

    def get_position(self):
        return (self.x, self.y)
```

```python
c = Circle(4, 0, 0)

print("Area:", c.area())
print("Perimeter:", c.perimeter())
print("Position:", c.get_position())
```

6. Implement a triangle class analogous to the rectangle and circle in question 5.

```python
class Triangle(Shape):
    def __init__(self, base, height, x, y):
        self.base = base
        self.height = height
        self.x = x
        self.y = y

    def area(self):
        return 0.5 * self.base * self.height

    def perimeter(self):
        # for a right angled triangle
        hyp = math.sqrt(self.base**2 + self.height**2)
        return self.base + self.height + hyp
```

```python
t = Triangle(3, 4, 0, 0)

print("Area:", t.area())
print("Perimeter:", t.perimeter())
```

8. Add a function to the object classes, including the base, that tests if a given set of $x$ and $y$ coordinates are inside of the object. You'll have to think through how to determine if a set of coordinates are inside an object for each object type.

```python
class Rectangle(Shape):
    def __init__(self, length, width, x, y):
        self.len = length
        self.w = width
        self.x = x
        self.y = y

    def area(self):
        return self.len * self.w

    def perimeter(self):
        return 2 * (self.len + self.w)
    
    def boundary_points(self):
        points = []
        for i in range(4):
            points.append((self.x + i, self.y))
        return points
    
    def contains(self, px, py):
        return (self.x <= px <= self.x + self.len and self.y <= py <= self.y + self.w)
```

```python
class Circle(Shape):
    def __init__(self, radius, x, y):
        self.rad = radius
        self.x = x
        self.y = y

    def area(self):
        return math.pi * self.rad ** 2

    def perimeter(self):
        return 2 * math.pi * self.rad
    
    def contains(self, px, py):
        dx = px - self.x
        dy = py - self.y
        return dx*dx + dy*dy <= self.rad**2
```

```python
class Canvas:
    def __init__(self, width, height):
        self.width = width
        self.height = height
        self.data = [[' '] * width for _ in range(height)]

    def set_pixel(self, row, col, char='*'):
        if 0 <= row < self.height and 0 <= col < self.width:
            self.data[row][col] = char

    def display(self):
        for row in self.data:
            print("".join(row))
```

```python
class Circle(Shape):
    def __init__(self, radius, x, y):
        self.rad = radius
        self.x = x
        self.y = y

    def area(self):
        return math.pi * self.rad ** 2

    def perimeter(self):
        return 2 * math.pi * self.rad
    
    def contains(self, px, py):
        dx = px - self.x
        dy = py - self.y
        return dx*dx + dy*dy <= self.radius**2


    def __repr__(self):
        return f"Circle({self.rad}, {self.x}, {self.y})"
```

```python
class RasterDrawing:
    def __init__(self):
        self.shapes = []

    def add(self, shape):
        self.shapes.append(shape)

    def paint(self):
        print("Painting shapes:")
        for shape in self.shapes:
            print("Area:", shape.area())

    def save(self, filename):
        with open(filename, "w") as f:
            for shape in self.shapes:
                f.write(repr(shape) + "\n")   
```

```python
def load_drawing(filename):
    drawing = RasterDrawing()

    with open(filename, "r") as f:
        for line in f:
            shape = eval(line.strip())
            drawing.add(shape)

    return drawing
```

```python
d = RasterDrawing()
d.add(Rectangle(10, 5, 2, 3))
d.add(Circle(4, 6, 6))

d.save("my_drawing.txt")

loaded_drawing = load_drawing("my_drawing.txt")

print("Loaded shapes:", loaded_drawing.shapes)
```
