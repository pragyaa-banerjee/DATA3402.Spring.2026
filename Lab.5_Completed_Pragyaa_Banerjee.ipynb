1. Create a `matrix` class with the following properties:
    * It can be initialized in 2 ways:
        1. with arguments `n` and `m`, the size of the matrix. A newly instanciated matrix will contain all zeros.
        2. with a list of lists of values. Note that since we are using lists of lists to implement matrices, it is possible that not all rows have the same number of columns. Test explicitly that the matrix is properly specified.
    * Matrix instances `M` can be indexed with `M[i][j]` and `M[i,j]`.
    * Matrix assignment works in 2 ways:
        1. If `M_1` and `M_2` are `matrix` instances `M_1=M_2` sets the values of `M_1` to those of `M_2`, if they are the same size. Error otherwise.
        2. In example above `M_2` can be a list of lists of correct size.


```python
class Matrix:
    
    def __init__(self, data, m=None):

        # CASE 1: Matrix(n, m) creates n x m filled with zeros.
        if isinstance(data, int) and isinstance(m, int):

            self.rows = data #stores number of rows
            self.cols = m #stores number of columns
            self.data = []

            # Create rows filled with zeros
            for i in range(self.rows):
                row = []
                for j in range(self.cols):
                    row.append(0)
                self.data.append(row)

        # CASE 2: Matrix(list_of_lists) creates matrix using given values.
        elif isinstance(data, list):

            # Checking that matrix is not empty
            if len(data) == 0:
                raise ValueError("Matrix cannot be empty")

            # Checking that all rows have same length
            first_row_length = len(data[0])

            for row in data:
                if len(row) != first_row_length:
                    raise ValueError("All rows must have same number of columns")

            self.rows = len(data)
            self.cols = first_row_length
            self.data = data

        else:
            raise ValueError("Invalid input to Matrix constructor")


    def __str__(self):
        result = ""
        for row in self.data:
            result += str(row) + "\n" #helps in printing the matrix nicely, each row appeas in its own line
        return result


    def __getitem__(self, key):

        # If using comma indexing like M[i,j]
        if isinstance(key, tuple):

            row_key, col_key = key

            # If both are integers we will return single number
            if isinstance(row_key, int) and isinstance(col_key, int):
                return self.data[row_key][col_key]

            # If slicing
            new_data = []

            rows = self.data[row_key]

            for row in rows:
                new_data.append(row[col_key])

            return Matrix(new_data)

        # If using M[i]
        else:
            return self.data[key]

    def __setitem__(self, key, value): #for M[i,j] = value

        if isinstance(key, tuple):
            i, j = key
            self.data[i][j] = value
        else:
            self.data[key] = value
            

    def assign(self, other):    # This method copies values from another matrix

        # Checking if other is Matrix
        if isinstance(other, Matrix):

            if self.rows != other.rows or self.cols != other.cols:
                raise ValueError("Matrices must have same size")

            for i in range(self.rows):
                for j in range(self.cols):
                    self.data[i][j] = other.data[i][j]

        # Checking if other is list of lists
        elif isinstance(other, list):

            if len(other) != self.rows or len(other[0]) != self.cols:
                raise ValueError("Dimensions must match")

            for i in range(self.rows):
                for j in range(self.cols):
                    self.data[i][j] = other[i][j]

        else:
            raise ValueError("Invalid assignment")

        
           
    def shape(self):  # Return number of rows and number of columns
        return (self.rows, self.cols)

    # to return transpose we flip rows and columns
    def transpose(self):

        new_data = []

        # Looping through columns first
        for j in range(self.cols):

            new_row = []

            # then looping through rows
            for i in range(self.rows):
                new_row.append(self.data[i][j])

            new_data.append(new_row)

        return Matrix(new_data)

    # Returning one row as matrix
    def row(self, n):
        return Matrix([self.data[n]])

    # Returning one column as matrix
    def column(self, n):

        new_data = []

        for i in range(self.rows):
            new_data.append([self.data[i][n]])

        return Matrix(new_data)

    # Returning the smaller sub-matrix
    def block(self, n0, n1, m0, m1):

        new_data = []

        for i in range(n0, n1):

            new_row = []

            for j in range(m0, m1):
                new_row.append(self.data[i][j])

            new_data.append(new_row)

        return Matrix(new_data)
    
```

3. Write functions that create special matrices (note these are standalone functions, not member functions of your `matrix` class):
    * `constant(n,m,c)`: returns a `n` by `m` matrix filled with floats of value `c`.
    * `zeros(n,m)` and `ones(n,m)`: return `n` by `m` matrices filled with floats of value `0` and `1`, respectively.
    * `eye(n)`: returns the n by n identity matrix.

```python
def constant(n, m, c): #n by m matrix with only 'c' as every element
    data = []
    for i in range(n): # Looping through each row
        row = []
        for j in range(m): # Loop through each column
            row.append(float(c))  # Appending float version of c
        data.append(row)
    return Matrix(data)

# zeros(n, m) creates an n by m matrix filled with 0.0
def zeros(n, m):
    return constant(n, m, 0.0)

# ones(n, m) creates an n by m matrix filled with 1.0
def ones(n, m):
    return constant(n, m, 1.0)

# eye(n) creates an n by n identity matrix.
def eye(n):
    data = []
    for i in range(n):
        row = []
        for j in range(n):
            if i == j: # this is when we are on diagonal
                row.append(1.0)
            else:
                row.append(0.0)

        data.append(row)

    return Matrix(data)
```

```python
class Matrix:
    
    def __init__(self, data, m=None):

        # CASE 1: Matrix(n, m) creates n x m filled with zeros.
        if isinstance(data, int) and isinstance(m, int):

            self.rows = data #stores number of rows
            self.cols = m #stores number of columns
            self.data = []

            # Create rows filled with zeros
            for i in range(self.rows):
                row = []
                for j in range(self.cols):
                    row.append(0)
                self.data.append(row)

        # CASE 2: Matrix(list_of_lists) creates matrix using given values.
        elif isinstance(data, list):

            # Checking that matrix is not empty
            if len(data) == 0:
                raise ValueError("Matrix cannot be empty")

            # Checking that all rows have same length
            first_row_length = len(data[0])

            for row in data:
                if len(row) != first_row_length:
                    raise ValueError("All rows must have same number of columns")

            self.rows = len(data)
            self.cols = first_row_length
            self.data = data

        else:
            raise ValueError("Invalid input to Matrix constructor")


    def __str__(self):
        result = ""
        for row in self.data:
            result += str(row) + "\n" #helps in printing the matrix nicely, each row appeas in its own line
        return result


    def __getitem__(self, key):

        # If using comma indexing like M[i,j]
        if isinstance(key, tuple):

            row_key, col_key = key

            # If both are integers we will return single number
            if isinstance(row_key, int) and isinstance(col_key, int):
                return self.data[row_key][col_key]

            # If slicing
            new_data = []

            rows = self.data[row_key]

            for row in rows:
                new_data.append(row[col_key])

            return Matrix(new_data)

        # If using M[i]
        else:
            return self.data[key]

    def __setitem__(self, key, value): #for M[i,j] = value

        if isinstance(key, tuple):
            i, j = key
            self.data[i][j] = value
        else:
            self.data[key] = value
            

    def assign(self, other):    # This method copies values from another matrix

        # Checking if other is Matrix
        if isinstance(other, Matrix):

            if self.rows != other.rows or self.cols != other.cols:
                raise ValueError("Matrices must have same size")

            for i in range(self.rows):
                for j in range(self.cols):
                    self.data[i][j] = other.data[i][j]

        # Checking if other is list of lists
        elif isinstance(other, list):

            if len(other) != self.rows or len(other[0]) != self.cols:
                raise ValueError("Dimensions must match")

            for i in range(self.rows):
                for j in range(self.cols):
                    self.data[i][j] = other[i][j]

        else:
            raise ValueError("Invalid assignment")

        
           
    def shape(self):  # Return number of rows and number of columns
        return (self.rows, self.cols)

    # to return transpose we flip rows and columns
    def transpose(self):

        new_data = []

        # Looping through columns first
        for j in range(self.cols):

            new_row = []

            # then looping through rows
            for i in range(self.rows):
                new_row.append(self.data[i][j])

            new_data.append(new_row)

        return Matrix(new_data)

    # Returning one row as matrix
    def row(self, n):
        return Matrix([self.data[n]])

    # Returning one column as matrix
    def column(self, n):

        new_data = []

        for i in range(self.rows):
            new_data.append([self.data[i][n]])

        return Matrix(new_data)

    # Returning the smaller sub-matrix
    def block(self, n0, n1, m0, m1):

        new_data = []

        for i in range(n0, n1):

            new_row = []

            for j in range(m0, m1):
                new_row.append(self.data[i][j])

            new_data.append(new_row)

        return Matrix(new_data)
    
    # to multiply every element in matrix by scalar c
    def scalarmul(self, c):

        new_data = []

        # Looping through rows
        for i in range(self.rows):

            new_row = []

            # Looping through columns
            for j in range(self.cols):

                new_row.append(self.data[i][j] * c) # here we multiply element by scalar

            new_data.append(new_row)

        return Matrix(new_data)
    
    # Adding two matrices element-wise
    def add(self, N):
        if self.shape() != N.shape():  # Checking if dimensions match
            raise ValueError("Matrices must be same size")

        new_data = []

        for i in range(self.rows):

            new_row = []

            for j in range(self.cols):

                new_row.append(self.data[i][j] + N.data[i][j]) # Adding corresponding elements

            new_data.append(new_row)

        return Matrix(new_data)
    
       
    # to subtract two matrices element-by-element
    def sub(self, N):

        if self.shape() != N.shape():
            raise ValueError("Matrices must be same size")

        new_data = []

        for i in range(self.rows):

            new_row = []

            for j in range(self.cols):

                new_row.append(self.data[i][j] - N.data[i][j])

            new_data.append(new_row)

        return Matrix(new_data)
    
    # Matrix multiplication
    def mat_mult(self, N):

        # Checking multiplication compatibility, if A is (a x b) and B is (b x c) then AB is (a x c).
        if self.cols != N.rows:
            raise ValueError("Invalid dimensions for multiplication")

        new_data = []

        # Looping through rows of first matrix
        for i in range(self.rows):

            new_row = []

            # Looping through columns of second matrix
            for j in range(N.cols):
                total = 0
                for k in range(self.cols):
                    total += self.data[i][k] * N.data[k][j]  # Dot product calculation

                new_row.append(total)

            new_data.append(new_row)

        return Matrix(new_data)
    
   
    def element_mult(self, N):

        if self.shape() != N.shape():
            raise ValueError("Matrices must be same size")

        new_data = []

        for i in range(self.rows):

            new_row = []

            for j in range(self.cols):

                new_row.append(self.data[i][j] * N.data[i][j]) # Multiplying corresponding elements of matrices

            new_data.append(new_row)

        return Matrix(new_data)

    
    #Here we check if two matrices are exactly equal
    def equals(self, N):

        if self.shape() != N.shape(): #First we check of the two matrices have the same dimensions
            return False

        for i in range(self.rows):
            for j in range(self.cols):
                if self.data[i][j] != N.data[i][j]:  #here we compare the matrices element-wise
                    return False

        return True
        
        
        
        
#adding the new functions:
       
    def __add__(self, other): # Overloading + operator
        return self.add(other)

    def __sub__(self, other): # Overloading - operator
        return self.sub(other)

    def __mul__(self, other): # Overloading * operator

        # If multiplying by matrix
        if isinstance(other, Matrix):
            return self.mat_mult(other)

        # If multiplying by scalar- for (Matrix*scalar )
        return self.scalarmul(other)

    # for (scalar * Matrix)
    def __rmul__(self, other):
        return self.scalarmul(other)

    def __eq__(self, other):# Overloading == operator
        return self.equals(other)
```
